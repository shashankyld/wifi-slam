# Fingerprint Similarity

```python
def overlap_APs(fi,fj):
    fi = np.array(fi)
    fj = np.array(fj)
    fi_mask = (fi < 0.0)
    fj_mask = (fj < 0.0)
    overlap_array_mask = np.array((fi_mask & fj_mask))
    return overlap_array_mask
```

```python
def detection_likelihood(fi, fj):
    fi = np.array(fi)
    fj = np.array(fj)
    fi_mask = (fi < 0.0)
    fj_mask = (fj < 0.0)
    Li = np.sum(fi_mask)
    Lj = np.sum(fj_mask)
    H = np.sum(overlap_APs(fi,fj))
    dl = np.divide(H, Li + Lj - H)
    return dl
```

```python
def signal_strength_likelihood(fi, fj):
    sigma = 0.5
    overlap_array_mask = overlap_APs(fi, fj)
    overlap_indices = np.where(overlap_array_mask)[0]
    H = np.sum(overlap_array_mask)
    prod = 1.0
    for idx in overlap_indices:
        a = np.power(fi[idx] - fj[idx], 2)
        b = np.divide(a, 2*np.power(sigma,2))
        c = np.exp(-b)
        prod = prod * c
    
    prod = np.power(prod, np.divide(1,H))
    #ssl = prod ** (np.divide(1,H))
    ssl = prod
    return ssl
```

```python
def similarity_score(fi, fj):
    dl = detection_likelihood(fi, fj)
    ssl = signal_strength_likelihood(fi, fj)
    sim = dl * ssl
    
    return sim
```

### Loop Closure Part

#### Loop Closure Conditions:

1. FP similarity between two positions is greater than certain threshold
2. Difference of cummulative distance (i.e. distance traveled so far) between two positions is greater than certain threshold

#### Implementation

```python
# Window size
w = 6
```

```python
def find_j_star(df, i, j, tau, w):
    den = 0.0
    num = np.array([0,0])
    
    # Find similarities between x_i_tau and all points in x_j sequence (-w/2,w/2 sequence around j)
    similarities = []
    indices = []
    for a in range(int(-w/2),int(w/2) + 1):
        sim_itau_l = similarity_score(df.iloc[(i+tau) % len(df), [9,10,11,12,13,14]], 
                                      df.iloc[(j+a) % len(df), [9,10,11,12,13,14]])
        
        similarities.append(sim_itau_l)
        indices.append((j+a) % len(df))
    
    # Take k nearest neighbours in similarity space, and calculate x_j_star
    k = 1
    similarities = np.array(similarities)
    indices = np.array(indices)
    
    data = np.vstack([similarities, indices])

    # Sort the indices of the first row in descending order
    sorted_indices = np.argsort(data[0])[::-1]

    # Select the top k indices
    top_indices = sorted_indices[:k]

    # Use fancy indexing to extract the columns
    result_array = data[:, top_indices]
        
    # Calculate x_j star now, since I have k highest similarities and indices of pooints from df
    den = 0.0
    num = np.array([0,0])
    
    for i in range(0, len(result_array[0])):
        den = den + result_array[0,i]
        num = num + (result_array[0,i] * df.iloc[int(result_array[1,i]) % len(df), [1,2]])
    
    j_star = np.divide(num, den)
    return j_star
```

```python
# Take xj_star itself
def find_correspondences(df, i, j, w):
    correspondences_points = []
    for tau in range(-int(w/2),int(w/2) + 1):
        correspondence = {}
        correspondence['first'] = list(df.iloc[(i+tau) % len(df), [1,2]])
        x_j_star = find_j_star(df, i, j, tau, w)
        correspondence['second'] = list(x_j_star)
        correspondences_points.append(correspondence)
    
    return correspondences_points
```

```python
def ICP_known_correspondences(df, correspondences):
    x_n = []
    y_n = []
    for k in range(0, len(correspondences)):
        x_n.append(correspondences[k]['first'])
        y_n.append(correspondences[k]['second'])
    
    
    x_n = np.array(x_n) # nx2
    y_n = np.array(y_n) # nx2
    
    x_n = x_n.transpose() # 2xn
    y_n = y_n.transpose() # 2xn
    
    # For checking -> seems to be correct
    #x_n = np.array([[0,0,0], [1,0,0],[0,1,0]]).transpose()
    #y_n = np.array([[0,0,1], [1,0,1],[0,0,2]]).transpose() #Approx transformation is 90 degree rot over x-axis and +1 in Z axis
    
    x_0 = np.divide(np.sum(x_n, axis=1), len(x_n)).reshape((-1,1)) # 2x1
    y_0 = np.divide(np.sum(y_n, axis=1), len(y_n)).reshape((-1,1)) # 2x1
    
    x_n_prime = x_n - x_0 # 2xn
    y_n_prime = y_n - y_0 # 2xn
    
    
    #Calculate covariance matrix (2x2)
    H = np.matmul(x_n_prime, y_n_prime.transpose())
    H = H.astype(np.float64)
    
    
    #Calculate singular value decomposition (SVD)
    U, D, V_T = np.linalg.svd(H) #the SVD of linalg gives you Vt
    D = np.diag(D)
    V = V_T.transpose()
    
    R = np.matmul(V, U.transpose())
    #R = np.matmul(U, np.transpose(V))
    t = y_0 - np.matmul(R,x_0)
    
    return R,t
```

```python
def ICP_error(df, correspondences, R, t):
    x_n = []
    y_n = []
    for k in range(0, len(correspondences)):
        x_n.append(correspondences[k]['first'])
        y_n.append(correspondences[k]['second'])
    
    x_n = np.array(x_n) # nx2
    y_n = np.array(y_n) # nx2
    
    x_n = x_n.transpose() # 2xn
    y_n = y_n.transpose() # 2xn
    
    error = 0.0
    for i in range(0, len(correspondences)):
        y_i_prime = np.matmul(R, x_n[:,i]) + t.flatten()
        error = error + np.linalg.norm(y_i_prime - y_n[:,i])
    
    error = np.divide(error, len(correspondences))
    
    return error
```

```python
LC_pairs = []
LC_transformations = []
errors = []
errors_percentage = []
for i in range(0, len(df)):
    flag = False
    for j in range(0, i):
        if((df.loc[i,'similarities'][j] > 0.75) and (np.fabs(df.loc[i, 'traveled_distance'] - df.loc[j, 'traveled_distance'])) > 20.0):
            correspondences_points = find_correspondences(df, i, j, w)
            R_star, t_star = ICP_known_correspondences(df, correspondences_points)
            if(ICP_error(df, correspondences_points, R_star, t_star) < 3.0):
                
                pair = [i,j]
                LC_pairs.append(pair)
                error_a = t_star.flatten()
                error_b = np.array(gt[j]) - np.array(gt[i])
                
                error = np.linalg.norm(error_a - error_b)
                error_percentage = np.divide(error, np.linalg.norm(error_b))
                
                #error = error_a - error_b
                errors.append(error)
                errors_percentage.append(error_percentage)
                #LC_transformations.append(T_star)
```
